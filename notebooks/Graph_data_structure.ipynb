# Graph data structure

### Definition

Graphs are collections of things and the relationships or connections between them. The data in a graph are called nodes or vertices. The connections between the nodes are called edges.

### Type

There are two major type of graph, directed and undirected. Directed graph edges have the specific direction on other hand undirected graph edges are bidirectional.

### Representing graph

* **adjacency list**

In adjacency list each edge maintain the unordered list of neighbour to which it is connected. In below example we represent undirected graph which is given on left hand side as the adjacency list on right hand side of the figure. 

adjacency_list.png

* **adjacency matrix**

In adjacency matrix each node in the graph is written as a row heading and a column heading. If the connection exist between the pair of node, then it's weight can be indicated at the intersection of respective row and column (if weight are not specified in graph, it is set to 1 by default). In the below matrix on the left hand side weight of 1 represent connection and weight of 0 represent no connection. In undirected graph there is symmetry along the diagonal of matrix, this symmetry is not present in directed graph. 

Note:  the order in which you arrange nodes in column and row doesn't matter.

### Why we need to learn graph data structure in AI ?

Knowladge of graph data structure is prerequisite for learning searching problem in AI. Algorithm such as depth-First search, breadth-First etc. used graph to traversing.
