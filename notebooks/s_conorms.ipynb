```
#import required library 
import numpy as np
```

### maximum operation

```
def s_max(x,y):
    return max(x,y)

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = s_max(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### algebric sum

```
def s_al(x,y):
    return x + y -x*y

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = s_al(i,j)
    output.append(result)


print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### bounded sum 

```
def s_bs(x,y):
    return min(1,x+y)

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = s_bs(i,j)
    output.append(result)


print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### drastic sum

```
def s_ds(x,y):
    if y == 0:
        return x
    if x == 0:
        return y
    if x>0 or y>0:
        return 0

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = s_ds(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```
