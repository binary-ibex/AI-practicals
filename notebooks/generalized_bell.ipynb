# Generalized bell-shaped membership function

$
f(x,[a,b,c]) = \Large\frac{1}{1 + |\frac{x-c}{a}|^{2b}}
$

bell_function.jpg

```
#import required library
import matplotlib.pyplot as plt 
import numpy as np 
```

```
#Generalized bell-shaped membership function
def gbellmf(x, params):
    a,b,c = params
    return 1/(1 + abs((x-c)/a)**(2*b))
```

```
#here x is the input series and y is output series after applying funtion to each value in x 
x = np.arange(0,10,0.1)
y = gbellmf(x, [2,4,6])
```

```
#plot the output 
plt.plot(x,y)
plt.xlabel('x -->', fontsize='19')
plt.ylabel('y -->', fontsize='19')
plt.title('Generalized bell', fontsize='19')
plt.show()
```
