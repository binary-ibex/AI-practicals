### What is breadth first search ? 

The Breadth First Search is an algorithm for traversing or searching tree or graph data structures. It explores all the nodes at the present depth before moving on to the nodes at the next depth level.


Let's see how it works 

- We start from node 'A' and add it to the queue.

queue_0.jpg



- We pop node 'A' out of the queue and add to the visited list.  Then we put all of its neighbour in the queue, namely node 'B' and node 'C' 

queue_5.jpg

- We pop node 'B' out of queue and put it in visited list. We put all unvisited neighbour of node 'B' in the queue.

queue_1.jpg

- We pop node 'C' out of queue and put it in visited list. Since node 'A' is already visited, noting is added to queue.

queue_2.jpg

- We pop node 'D' out of queue and put it in visited list. Since all neighbour of node 'D' are visited, we move to node 'E'

queue_3.jpg

- We pop node 'E' out of queue and put it in visited list. Since all neighbour of node 'E' are visited, queue remains empty and graph traversing end.

### graph traversing using breadth first search 


```
def bfs(graph, start):
    queue=[]
    visited = []
    queue.append(start)
    while(queue):
        N = queue.pop(0)
        if N not in visited:
            visited.append(N)
        if not N:
            continue
        for node in graph[N]:
            if node not in visited:
                queue.append(node)
    return visited

graph = {
    'A' : ['B' , 'C'],
    'B' : ['D' , 'E'],
    'C' : ['A'],
    'D' : ['B'],
    'E' : ['B']
}

print ( bfs(graph, 'A') )
```

### using breadth first search to find the shortest find between node 'A' and 'F' in given graph.

breadth_first_graph.jpg

```
def find_path(graph, start, end) :

    #is start node does not exist in graph
    if start not in graph:
        return None

    #path
    path = []

    #if we find the path
    if start == end:
        return [start]

    queue=[]

    #list for keeping track of the visited nodes
    visited = []

    #dictionary from tracking previous node
    back_track = {}

    #adding start node to back track
    back_track[start] = None
    #add first element to the queue
    queue.append(start)



    while(queue):
        #poping node from queue
        N = queue.pop(0)

        #adding it to visited list
        if N not in visited:
            visited.append(N)

        #if leaf node then continue
        if not N:
            continue

        #looping through all of its neighbour
        for node in graph[N]:

            #if node is unvisited then add it to queue
            if node not in visited:
                queue.append(node)
                back_track[node] = N

            #if we find the destination, then reconstruct the path using back_track dictionary
            if node == end:
                while end != start:
                    path = [end] + path
                    end = back_track[end]
                path = [start] + path
                return path

    #if path does not exist return none 
    return None




if __name__=='__main__':

    graph = {
        'A':['B', 'C', 'D'],
        'B':['A' , 'D', 'E'],
        'C':['A', 'D'],
        'D':['B', 'A', 'C'],
        'E':['B', 'F'],
        'F':['E']
    }

print ( find_path(graph, 'A' , 'F'))
```
