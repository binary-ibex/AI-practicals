```
import numpy as np 
```

### Minimum operation 

```
def t_min(x,y):
    return min(x,y)


np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = t_min(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### algebric product

```
def al_p(x,y):
    return x*y


np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = al_p(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### bounded product

```
def b_p(x,y):
    return max(0, x+y-1)


np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = b_p(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### drastic product

```
def d_p(x,y):
    if y == 1:
        return x
    if x == 1:
        return y
    if x<1 or y<1:
        return 0

np.random.seed(200)   

a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = d_p(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```
