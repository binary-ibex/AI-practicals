```
import numpy as np 

#seed the random number
np.random.seed(10)

#input data
input_data = [
	[0,0],
	[0,1],
	[1,0],
	[1,1]
]

#actual output data
output = [
	[0],
	[0],
	[0],
	[1]
]

# convert input and output to numpy array 
input_data = np.array(input_data)
output = np.array(output)

# initalize weight and bias
weights = np.random.randn(2,1)
bias = np.random.randn(1)

#setup learning rate 
learning_rate = 0.08


#sigmoid activation function
def sigmoid(x):
	return 1/(1 + np.exp(-x))

#derivative of sigmoid activation
def der_sigmoid(x):
	return x*(1-x)

#train the network 
for _epoch in range(1000):

	#forward propagation 
	prediction = sigmoid(input_data.dot(weights) + bias)

	loss = 0.5*sum((prediction - output)**2)


	#back propagation
	alpha = (prediction - output)*der_sigmoid(prediction)
	#update weight 
	weights = weights - learning_rate*input_data.T.dot(alpha)
	#update bias 
	bias = bias -  learning_rate*sum(alpha)



	#print loss after every 100 epoch 
	if _epoch%100 == 0:
		print(f"epoch {_epoch} loss {loss[0]}")


#print the final output 
print("\nprediction\n")
print(sigmoid(input_data.dot(weights) + bias))
```

```
import numpy as np

#seed the random number
np.random.seed(10)

#input data
input_data = [
	[0,0],
	[0,1],
	[1,0],
	[1,1]
]

#actual output data
output = [
	[0],
	[1],
	[1],
	[0]
]

# convert input and output to numpy array
input_data = np.array(input_data)
output = np.array(output)

# initalize weight and bias
weight1 = np.random.randn(2,3)
weight2 = np.random.randn(3,1)
bias1 = np.random.randn(3)
bias2 = np.random.randn(1)

#setup learning rate
learning_rate = 0.4


#sigmoid activation function
def sigmoid(x):
	return 1/(1 + np.exp(-x))

#derivative of sigmoid activation
def der_sigmoid(x):
	return x*(1-x)

#train the network
for _epoch in range(5000):

	#forward propagation
	hidden_layer = sigmoid(input_data.dot(weight1) + bias1)
	prediction = sigmoid(hidden_layer.dot(weight2) + bias2)

	loss = 0.5*sum((prediction - output)**2)


	#Backpropagation
	alpha1 = (prediction - output) * der_sigmoid(prediction)

	#update weight2
	weight2 = weight2 - learning_rate*hidden_layer.T.dot(alpha1)

	#update bias2
	bias2 = bias2 - learning_rate*alpha1.sum(axis=0)

	alpha2 = alpha1*weight2.T*der_sigmoid(hidden_layer)

	#update weight1
	weight1 = weight1 - learning_rate*input_data.T.dot(alpha2)

	#update bias1
	bias1 = bias1 - learning_rate*alpha2.sum(axis=0)

	#print loss after every 100 epoch
	if _epoch%200 == 0:
		print(f"epoch {_epoch} loss {loss[0]}")


#print the final output after training
hidden_layer = sigmoid(input_data.dot(weight1) + bias1)
prediction = sigmoid(hidden_layer.dot(weight2) + bias2)
print("\nprediction\n")
print(prediction)
```
