# Implement adjacency list using python 

Adjacency list is one of the best way of representing graph data structure and python makes it very easy to implement it. We are going to use python dictonary to implement adjacency list. We can also use python class to implement it, but I found python dictonary much more simple to work with.

> **Note**: I am assuming you are already familiar with python dictionary. In case you don't, here is the best resource to learn about it =>  https://realpython.com/python-dicts/

Idea is very simple, we make the dictionary, in it we put every node in the graph as key and list of all neighbour as the corresponding value of key. Let's walk through few examples to understand this idea better. 

# Undirected Graph 

graph_1.jpg

```
graph = {
    'A' : ['B'],
    'B' : ['A']
}


print(graph)    #print's graph 
print(graph['A'])  #prints list of node connected to 'A'
print(graph['B'])  #prints list of node connected to 'B
```

graph_2.jpg

```
graph = {
    'A': ['B', 'D'],
    'B': ['A', 'C'],
    'C': ['B', 'D'],
    'D': ['A', 'C']
}

print(graph)
print(graph['A'])
print(graph['B'])
print(graph['C'])
print(graph['D'])
```

graph_3.jpg

```
graph = {
		'A' : ['B'], 
		'B' : ['D', 'C'],
		'C' : ['B', 'E'],
		'D' : ['B', 'F'],
		'E' : ['C'],
		'F' : ['D']
		}
print(graph)
print(graph['A'])
print(graph['B'])
print(graph['C'])
print(graph['D'])
print(graph['E'])
print(graph['F'])
```

```
graph = {
		'A' : ['B'], 
		'B' : []
		}
print(graph)
print(graph['A'])
print(graph['B'])
```

# Weighted graph

In the weighted graph we use the another dictonary as the value instead of list 

graph_1.jpg

```
graph = {
		'A' : {'B' : 10}, 
		'B' : {'A' : 10}
		}

print (graph)
print (graph['A'])
print (graph['B'])
```

graph_2.jpg

```
graph = {
		'A' : {'B' : 10}, 
		'B' : {}
		}

print (graph)
print (graph['A'])
print (graph['B'])
```

graph_3.jpg

```
graph = {
		'A' : {'B' : 10}, 
		'B' : {'A' : 10, 'C' : 20},
		'C' : {'B' : 20, 'E' : 30, 'D' : 60},
		'D' : {'C' : 60},
		'E' : {'C' : 30}
		}

print (graph)
print (graph['A'])
```

graph_5.jpg

```
graph = {
		'A' : {'B' : 10}, 
		'B' : {'C' : 20},
		'C' : {'D' : 30},
		'D' : {}
		}
print(graph)
```
