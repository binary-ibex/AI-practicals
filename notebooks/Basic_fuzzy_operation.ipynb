```
import numpy as np 
```

### Union 

```
def union(x,y):
    return max(x,y)

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = union(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### intersection


```
def intersection(x,y):
    return min(x,y)

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output = []

for i,j in zip(a,b):
    result = intersection(i,j)
    output.append(result)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output}")
```

### complement

```
def complement(x):
    return 1 - x

np.random.seed(200)
a = np.random.rand(5)
b = np.random.rand(5)

output_a = []
output_b = []

for i,j in zip(a,b):
    result_a = complement(i)
    result_b = complement(j)

    output_a.append(result_a)
    output_b.append(result_b)

print(f"a = {a}")
print(f"b = {b}")
print(f"output = {output_a}")
print(f"output = {output_b}")
```
