# Gradient descent 

consider equation 

$y = (x-3)^2+2$

graph.jpg

Find the value of x for which value of y=2

Ans. 

 1. Randomly initalized the value of x

    let $x = 2$
 2. predict the value of y 
    
    $prediction = (x - 3)^2 + 2 $

    $prediction = (2-3)^2 + 2 = 3$

 3. calculate loss

    $loss = \frac{1}{2}(prediction - y)^2$

    $loss = \frac{1}{2}(3-2)^2 = 0.5$

4. calculate derivative of loss w.r.t x

    $\frac{d \ loss}{dx} = \frac{d}{dx}(\frac{1}{2}(prediction - y)^2)$

    applying chain rule we get,

    $=(prediction - y)* \frac{d \ prediction}{dx}$

    $=(prediction - y) * 2*(x-3)$

    $= (3 - 2 )* 2 * ( 2 - 3) = -1$

5.  update the value of x using gradient descent

    $x = x - \alpha * \frac{d \ loss}{dx}$

    $x = 2 - 0.01 * (-1)$

    $x = 2.01$


6. Repeat process from step 2 to step 5 for number of time for every new value of x. 



```
#gradient descent 

# function  y = (x-3)^2 + 2
def fun(x):
    return (x-3)**2 + 2


#derivative of function 
def der_fun(x):
    return 2*(x-3)

# here I randomly initalize value of x =2 you can take any value 
x = 2

#actual value of y 
y = 2

#set learning rate 
learning_rate = 0.01


#inital values
print("\n")
print(f"Initial value of x = {x}")
print(f"Initial prediction  = {fun(x)}")
print(f"Initial loss = {0.5*((fun(x) - y )**2)}")
print("\n")

#loop to update the value of x 
for _iter in range(100):

    prediction = fun(x)

    loss = 0.5*((prediction - y )**2)

    x = x - learning_rate*(prediction - y)*der_fun(x)

    if _iter%5 == 0:
        print(f"value of x = {x}        loss = {loss}")



#final values 
print("\n")
print(f"final  value of x = {x}")
print(f"final value prediction  = {fun(x)}")
print(f"final value  loss = {0.5*((fun(x) - y )**2)}")
```

Notice how the value of the loss decrease at each iteration and value of x approch to the actual required value. Here I iterate 100 time, you can iterate more to further decrease the loss.
